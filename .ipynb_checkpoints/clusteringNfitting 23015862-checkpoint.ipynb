```python
# Select specific columns for clustering analysis
df_clust = sub_df[['Annual Income (k$)', 'Spending Score (1-100)']].copy()
# `df_clust` contains 'Annual Income' and 'Spending Score' from `sub_df`

# Initialize a RobustScaler for scaling the data
scaler = RobustScaler()  # RobustScaler is chosen for its resistance to outliers

# Fit and transform the data to normalize it
norm = scaler.fit_transform(df_clust)  # Normalize the data in `df_clust` using the RobustScaler

# After scaling, the data in `norm` is centered and scaled to minimize the influence of outliers
```

```python
def compute_silhouette_inertia(n, data):
    """
    Compute the silhouette score and within-cluster sum of squares (WCSS) for a given number of clusters.
    """
    # Create a KMeans instance with the specified number of clusters
    clusterer = KMeans(n_clusters=n, n_init=20)  # Initialize KMeans with n clusters and 20 restarts
    
    # Fit the KMeans model to the data
    clusterer.fit(data)  # Apply KMeans clustering to the input data
    
    # Get the cluster labels assigned to each data point
    cluster_labels = clusterer.labels_  # Retrieve the labels for each data point
    
    # Compute the silhouette score to evaluate the clustering quality
    sil_score = silhouette_score(data, cluster_labels)  # Calculate silhouette score
    
    # Retrieve the within-cluster sum of squares (WCSS)
    wcss = clusterer.inertia_  # WCSS indicates the tightness of clusters
    
    return sil_score, wcss  # Return both silhouette score and WCSS


#(Cooper, 2024)
```

```python
def elbow_plot(min_k, max_k, wcss, optimal_k):
    """
    Plots the Elbow Method to determine the optimal number of clusters
    based on the WCSS metric.
    """
    # Create a figure and axis with a higher DPI for better quality
    fig, ax = plt.subplots(dpi=144)
    
    # Plot WCSS values as a line
    ax.plot(range(min_k, max_k + 1), wcss, linestyle='-', marker='x', color='black')
    
    # Highlight the optimal number of clusters with a red circle
    ax.scatter(optimal_k, wcss[optimal_k - min_k], s=50, color='red', edgecolors='black', label=f"Optimal k = {optimal_k}")
    
    # Label the axes and add a legend
    ax.set_xlabel('Number of Clusters (k)')
    ax.set_ylabel('Within-Cluster Sum of Squares (WCSS)')
    ax.set_title('Elbow Method to Determine Optimal Clusters')
    
    # Add limits to the x-axis for clarity
    ax.set_xlim(min_k, max_k)
    
    # Save the plot to a PNG file and display it
    plt.savefig('figure-3.png')
    plt.show()
    
    return


#(Cooper, 2024)
```

```python
wcss = []
best_n, best_score = None, -np.inf
for n in range(2, 11):  # 2 to 10 clusters
    score, inertia = compute_silhouette_inertia(n, norm)
    wcss.append(inertia)
    if score > best_score:
        best_n = n
        best_score = score
    print(f"{n:2g} clusters silhoutte score = {score:0.2f}")

print(f"Best number of clusters = {best_n:2g}")
elbow_plot(2, 10, wcss, best_n)
```

```python
# Create a scatter plot with fitted line and error margins
def scatter_plot_with_fitted_line(x_data, y_data, x_fitted, y_fitted):
    """
    Plots a scatter plot with fitted line and error margins.
    """
    plt.figure(dpi=144)
    
    # Scatter plot for original data points
    plt.scatter(x_data, y_data, color='blue', label='Original Data')
    
    # Overlaid line representing the fit
    plt.plot(x_fitted, y_fitted, 'k--', label='Fitted Line (Scipy)')
    
    # Generate a line for the fitted model
    x_fit = np.linspace(np.min(x_data), np.max(x_data), 100)
    y_fit = linear_model(x_fit, params[0], params[1])
    plt.plot(x_fit, y_fit, 'k-', label='Fitted Line (Poly)')

    # Error margins for the fitted line
    plt.fill_between(x_fit, linear_model(x_fit, params[0] - errors[0], params[1] - errors[1]),
                     linear_model(x_fit, params[0] + errors[0], params[1] + errors[1]), 
                     color='gray', alpha=0.1, label='Error Margins')

    # Formatting the plot
    plt.xlabel('Annual Income (k$)')
    plt.ylabel('Spending Score (1-100)')
    plt.title('Scatter Plot with Fitted Line and Error Margins')
    plt.legend()
    plt.savefig('figure-5.png')
    plt.show()

# Independent variable data
x_income = sub_df['Annual Income (k$)']
# Dependent variable data
y_score = sub_df['Spending Score (1-100)']

# Define a linear model for curve fitting
def linear_model(x, slope, intercept):
    """
    Linear function to fit a straight line. 
    """
    return slope * x + intercept

# Fit the linear model to the data
params, covariance = curve_fit(linear_model, x_income, y_score)

# Calculate the errors from the covariance matrix
errors = np.sqrt(np.diag(covariance))

print(f"Slope = {params[0]:.2f} +/- {errors[0]:.2f}")
print(f"Intercept = {params[1]:.2f} +/- {errors[1]:.2f}")

# Using numpy to fit the data with a first-degree polynomial
poly_fit = Poly.fit(x_income, y_score, 1)

# Get the coefficients in reverse order from numpy
coefficients = poly_fit.convert().coef
print(f"Intercept = {coefficients[1]:.2f} +/- {errors[1]:.2f}")
print(f"Slope = {coefficients[0]:.2f} +/- {errors[0]:.2f}")

# Use a new set of fitted data
x_line = np.linspace(np.min(x_income), np.max(x_income), 100)
y_line = linear_model(x_line, params[0], params[1])

# Plot the data with fitted lines and error margins
scatter_plot_with_fitted_line(x_income, y_score, x_line, y_line)
```
